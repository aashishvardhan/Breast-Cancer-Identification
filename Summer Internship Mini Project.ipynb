# Breast Cancer Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as spy
```

```python
df = pd.read_csv ('breast-cancer-wisconsin.csv',
                  names= ['Sample code number','Clump Thickness','Uniformity of Cell Size',
                         'Uniformity of Cell Shape','Marginal Adhesion',
                         'Single Epithelial Cell Size','Bare Nuclei','Bland Chromatin',
                         'Normal Nucleoli','Mitoses','Class'])
df.head(10)
```

### Codition of Cancer

```python
plt.hist(df['Class'],color='y')
plt.title('Plot Diagnosis (Benign=2,Malignant=4)')
Benign, Malignant = df['Class'].value_counts()
print('Number of Benign: ', Benign)
print('Number of Malignant: ', Malignant)
plt.show()
```

```python
corr = df.corr() # Finding out the correlation between the features
sns.set(rc={'figure.figsize': (15,10)}) # Plotting the heatmap of correlation between features
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
sns.pairplot(df)
```

```python
x=df.drop(['Sample code number','Class'], axis=1)
x.head(5)
```

```python
y=df['Class']
y.head(5)
```

```python
y=df.replace(2,0,inplace=True)
y=df.replace(4,1,inplace=True)
```

```python
y=df['Class']
y.tail(5)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=11)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier() 
from sklearn.model_selection import GridSearchCV
k_range= list(range(1,31)) 
param_grid= dict(n_neighbors= k_range)
grid= GridSearchCV(knn, param_grid,cv=10,scoring='accuracy',return_train_score=False,verbose =1)
grid_search= grid.fit(xtrain, ytrain)
```

```python
print(grid_search.best_params_)
```

```python
accuracy=grid_search.best_score_
print("Results by KNN algorithm, is :", accuracy*100)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression 
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier 
from sklearn.svm import SVC

log= LogisticRegression()
linear = LinearRegression() 
dtc= DecisionTreeClassifier()
rfc = RandomForestClassifier() 
model_log= log.fit(xtrain, ytrain)
model_lin =linear.fit(xtrain, ytrain) 
model_dtc = dtc.fit(xtrain, ytrain)
model_rfc=rfc.fit(xtrain, ytrain)
```

```python
print("Result by using logisticRegression : ", round(100*model_log.score(xtest,ytest), ndigits = 4)) 
print("Result by using linearRegression : ", round(100*model_lin.score(xtest,ytest), ndigits = 4)) 
print("Result by using DecisionTreeClassifier: ",round(100*model_dtc.score(xtest,ytest), ndigits = 4))
print("Result by using RandomForestClassifier : ",round(100*model_rfc.score(xtest,ytest), ndigits = 4))
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
from sklearn.metrics._plot.confusion_matrix import plot_confusion_matrix 
plot_confusion_matrix (model_log,xtest,ytest, values_format = 'd')
```

## Accuracy

#### Logistic Regression 

```python
y_test=log.predict(xtest)
accuracy_score(ytest,y_test)
```

#### Random Forest classification

```python
y_test=rfc.predict(xtest)
accuracy_score(ytest,y_test)
```
