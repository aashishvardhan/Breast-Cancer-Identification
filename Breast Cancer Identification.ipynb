# Breast Cancer Data Analysis and Classification

### Columns Description

**sample code number** - a unique identifier for each medical test report<br>
**clump thickness** - the depth of a cluster or mass of cancer cells in the breast tissue<br>
**uniformity of cell size** - refers to how similar or different the cancer cells appear in terms of their size.<br>
**uniformity of cell shape** - refers to how similar or different the cancer cells appear in terms of their shape.<br>
**marginal adhesion** - the degree of attachment between cancer cells and the surrounding tissue.<br>
**single epithelial cell size** - the dimensions of the individual cells that make up the cancer tissue.<br>
**bare nuclei** - the absence of cytoplasm around the nucleus of a cell<br>
**bland chromatin** - the appearance of the chromatin (the material that makes up the nucleus of a cell) in cancer cells under a microscope.<br>
**normal nuclei** - presence of nuclei in cancer cells that have a similar appearance to the nuclei of normal, healthy breast cells.<br>
**mitoses** – gives information whether the cancer cells are being split into multiple daughter cells.<br>
**class** – tells us whether the patient is benign or malignant to breast cancer.<br>

```python
x=df.drop(['Sample code number','Class'], axis=1)
x.head(5)
```

```python
y=df['Class']
y.head(5)
```

```python
y=df.replace(2,0,inplace=True)
y=df.replace(4,1,inplace=True)
```

```python
y=df['Class']
y.tail(5)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=11)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier() 
from sklearn.model_selection import GridSearchCV
k_range= list(range(1,31)) 
param_grid= dict(n_neighbors= k_range)
grid= GridSearchCV(knn, param_grid,cv=10,scoring='accuracy',return_train_score=False,verbose =1)
grid_search= grid.fit(xtrain, ytrain)
```

```python
print(grid_search.best_params_)
```

```python
accuracy=grid_search.best_score_
print("Results by KNN algorithm, is :", accuracy*100)
```

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
model_lin = linear.fit(xtrain, ytrain)
```

```python
print("Accuracy of linearRegression : ", round(100*model_lin.score(xtest,ytest), ndigits = 4))
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
model_log = log.fit(xtrain, ytrain)
```

```python
print("Accuracy of logisticRegression : ", round(100*model_log.score(xtest,ytest), ndigits = 4)) 
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
model_dtc = dtc.fit(xtrain, ytrain)
```

```python
print("Accuracy of DecisionTreeClassifier: ",round(100*model_dtc.score(xtest,ytest), ndigits = 4))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
model_rfc = rfc.fit(xtrain, ytrain)
```

```python
print("Accuracy of RandomForestClassifier : ",round(100*model_rfc.score(xtest,ytest), ndigits = 4))
```

## Confusion Matrix of Top 2 Models

```python
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
```

#### Logistic Regression 

```python
y_test=log.predict(xtest)
accuracy_score(ytest,y_test)
```

```python
cm1 = confusion_matrix(ytest,y_test)
```

```python
cmv1 = ConfusionMatrixDisplay(cm1)
cmv1.plot()
```

#### Random Forest classification

```python
y_test=rfc.predict(xtest)
accuracy_score(ytest, y_test)
```

```python
cm2 = confusion_matrix(ytest, y_test)
```

```python
cmv2 = ConfusionMatrixDisplay(cm2)
cmv2.plot()
```
